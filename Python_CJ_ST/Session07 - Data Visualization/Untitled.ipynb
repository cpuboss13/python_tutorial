```python
# 커맨드뷰에서 차트 시여
%matplotlib inline
```

```python
x=list(range(0,100,5))
```

```python
x
```

```python
y=[]
for i in x:
    y.append(i**2)
```

```python
y
```

```python
plt.plot (x,y, 'm', lw=1.5)
plt.plot (x,y, 'o')
plt.grid (True)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Simple plot')
```

```python
%matplotlib inline
y=np.random.standard_normal((20,2))
y[:,1] = y[:, 1]*100

#x=list(range(0,100,5))
#y=[i**2 for i in x]
#y2=[i**2.2 for i in x]
plt.figure(figsize=(7,5))
plt.subplot(121)
plt.plot(y[:,0], lw=1.5, label='1st')
plt.plot(y[:,0],'o')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('A Simple Plot')
plt.subplot(122)
plt.plot(y[:,1], lw=1.5, label='2nd')
plt.plot(y[:,1],'o')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
       
```

```python
y=np.random.standard_normal((50,2))
c=np.random.randint(0,10,len(y))

plt.figure(figsize=(7,5))
plt.scatter(y[:,0],y[:,1],marker='o',c=c)
plt.colorbar()

plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter plt')

```

```python
import seaborn as sns
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.pairplot(tips)
sns.pairplot(tips, hue='sex')
```

```python
sns.Implot(x='total_bill', y='tip', data=tips)
sns.Implot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'])
```
